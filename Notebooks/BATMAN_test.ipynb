```python
import batman
import numpy as np
import matplotlib.pyplot as plt
```

```python
params = batman.TransitParams()
params.t0 = 0. #Transit time of inferior conjunction
params.per = 1. #Orbital period in days
params.rp = 0.1 #planet radius as a fraction of star radius
params.a = 15. #semi-major axis as a fraction of star radius
params.inc = 87. #orbital inclination (degrees)
params.ecc =0. #eccentricity
params.w = 90. #longitude of periastron (deg)
params.u = [0.1, 0.3] #limb darkening coefficients
params.limb_dark = 'quadratic' #limb darkening model
```

```python
t = np.linspace(-0.05, 0.05, 100)
```

```python
m = batman.TransitModel(params, t) #initialises model
flux = m.light_curve(params) #calculates light curve
```

```python
plt.plot(t,flux)
plt.xlabel('Time from central transit')
plt.ylabel('Relative flux')
plt.show()
```
